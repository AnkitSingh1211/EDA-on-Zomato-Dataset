```python
##Numerical variable
df_final.columns
```

```python
ratings=df_final.groupby(['Aggregate rating','Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating_count'})
```

```python
ratings
```

## Observation
1. When rating is between 4.5 to 4.9 --->Excellent
2. When rating is between 4.0 to 4.4 --->Very Good
3. When rating is between 3.5 to 3.9 --->Good
4. When rating is between 3.0 to 3.4 --->Average
5. When rating is between 2.5 to 2.9 --->Average

```python
plt.figure(figsize=(10,5))

sns.barplot(x='Aggregate rating',y='Rating_count',data=ratings,hue='Rating color',palette=['black','red','orange','yellow','green','green'])
```

Observation:
   1. Not rated count is very high
   2. Maximum number of rating are between 2.5 - 3.4

```python
##count plot 
sns.countplot(x='Rating color',data=ratings,palette=['black','red','orange','yellow','green','green'])
```
